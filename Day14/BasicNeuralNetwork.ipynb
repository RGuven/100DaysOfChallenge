```python
import numpy 
from numpy import exp, array, random, dot, mean, abs
```

```python
my_input = array([[0, 0, 1], [1, 1, 1], [1, 0, 1]]) 


my_output = array([[0, 1, 1]]).T
                    
weights = array([[1.0, 1.0, 1.0]]).T 
                
for loop in range(100000):
    neuron_value = dot(my_input,weights)
    predict = 1 / (1 + exp(-neuron_value)) 
    weights = weights + dot(my_input.T, ((my_output - predict) * predict * (1 - predict)))
    print(str(numpy.mean(numpy.abs(my_output-predict))))
 

print(1/(1+exp(-(dot(array([1,0,0]),weights)))))  # Actually my result is 1 
```
